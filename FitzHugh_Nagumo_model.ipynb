Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Initializing the parameters.

```python
a = 0.7
b = 0.8
tau = 12.5
R = 0.1
I_ext = 0.5
```

Model Equations

```python
def eqn_w(v,w):
    dwdt = v + a - (b*w)
    return dwdt*tau

def eqn_v(v,w):
    dvdt = v - ((v**3)/3) - w + I_ext
    return dvdt
```

Creating a Grid

```python
v = np.linspace(-2.5,2.5,400)
w = np.linspace(-1,2,400)
V,W = np.meshgrid(v,w)
```

```python
dV = eqn_v(V,W)
dW = eqn_w(V,W)
```

```python
plt.style.use('dark_background')

plt.figure(figsize = (8,6))
plt.streamplot(V,W,dV,dW, color='#EDC8FF')
```

```python
v_null = v - ((v**3)/3) + I_ext
w_null = (v + a)/b
```

Plotting the graphs

```python
plt.plot(v, v_null, label = 'v-nullcline',color = 'lightpink')
plt.plot(w, w_null, label = 'w-nullcline',color = '#EFBF04')

plt.xlabel('v (membrane potential)')
plt.ylabel('w (recovery variable)')
plt.title('Phase Plane Portrait')
plt.show()
```
